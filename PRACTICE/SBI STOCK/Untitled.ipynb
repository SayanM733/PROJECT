```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv(r'E:\ML\PROJECT\PRACTICE\SBI STOCK\SBI_BANK.csv')
df.dropna(inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
# Define the features and target variable
features = ['Open', 'Volume', 'VWAP']  # Remove inner list
X = df[features]
y = df['Close']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the RandomForestRegressor with corrected parameters
sbi = KNeighborsRegressor(n_neighbors=1)

# Train the model
sbi.fit(X,y)

# Make predictions
y_pred = sbi.predict([[577.750000,11777454,573.25]])
print(y_pred)
print("Predicted Close Price:", y_pred[0])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# Define the features and target variable
features = ['Open', 'Volume', 'VWAP']  # Remove inner list
X = df[features]
y = df['Close']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the RandomForestRegressor with corrected parameters
sbi = RandomForestRegressor(n_estimators=30, random_state=42)

sbi.fit(X,y)

# Make predictions
y_pred = sbi.predict([[577.750000,11777454,573.25]])
print(y_pred)
print("Predicted Close Price:", y_pred[0])
```
